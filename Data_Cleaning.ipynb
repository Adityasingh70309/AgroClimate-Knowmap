```python
import pandas as pd
```

```python
df = pd.read_csv("Downloads/archive (3)/crop_yield.csv")
```

```python
df['Crop'] = df['Crop'].str.strip()
df['Season'] = df['Season'].str.strip()
df['State'] = df['State'].str.strip()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Crop_Year'] = df['Crop_Year'].astype(int)
df['Area'] = df['Area'].astype(float)
df['Production'] = df['Production'].astype(float)
```

```python
df = df[(df['Area'] > 0) & (df['Production'] >= 0) & (df['Yield'] >= 0)]
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
print("âœ… Cleaned Dataset Summary:")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nSample Data:\n", df.head())
```

```python
df.to_csv("cleaned_crop_yield.csv", index=False)
print("\nðŸ’¾ Cleaned file saved as 'cleaned_crop_yield.csv'")
```
